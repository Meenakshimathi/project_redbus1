```python
if 'Price' in df.columns:
    # Clean the 'Price' column: Remove 'INR ' and handle NaN values
    df['Price'] = df['Price'].fillna('').astype(str).str.replace('INR ', '', regex=False)
    
    # Convert the 'Price' column to numeric (ignore errors in case of non-numeric values)
    df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
    
    # Sort the DataFrame by 'Price' in ascending order
    df = df.sort_values(by='Price', ascending=True)
```
